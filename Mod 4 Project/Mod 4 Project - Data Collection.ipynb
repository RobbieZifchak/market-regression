```python
import datetime
import itertools
```

```python
def date_range(start_date, end_date):
    for ordinal in range(start_date.toordinal(), end_date.toordinal()):
        yield datetime.date.fromordinal(ordinal)
```

```python
start = pd.to_datetime('2015-1-1')
end = pd.to_datetime('2019-12-31')
```

```python
all_dates = pd.DataFrame((date_range(start, end)))
```

# FOREX (US to EURO)

```python
forex_url = f'https://www.alphavantage.co/query?function=FX_DAILY&from_symbol=EUR&to_symbol=USD&outputsize=full&apikey=2SQ8ZIN2ZZ2EQZ91'
```

```python
def forex_api_call(fromtick,totick):
    url = f'https://www.alphavantage.co/query?function=FX_DAILY&from_symbol={fromtick}&to_symbol={totick}&outputsize=full&apikey=2SQ8ZIN2ZZ2EQZ91'  
    response = requests.get(url)
    
    data = response.text
    data_dict=json.loads(data)
    return data_dict
```

```python
fromtick= 'USD'
totick='EUR'
```

```python
forex_raw_data=forex_api_call(fromtick,totick)
```

FOREX Parse

```python
def forex_parse(data):

    forex = pd.DataFrame(forex_raw_data['Time Series FX (Daily)'])
    forex = forex.transpose()
    forex = forex.reset_index()
    forex=forex.drop(columns = ['1. open', '2. high','3. low'])
    forex= forex.rename(columns={'index': "Date", '4. close': 'Value'})
    forex['Value']= forex['Value'].apply(pd.to_numeric)
    

    return forex
```

```python
forex = forex_parse(forex_raw_data)
```

```python
forex['Date']=pd.to_datetime(forex['Date'])
```

```python
forex = forex[forex['Date'] < '2019-12-31']
forex = forex[forex['Date'] > '2014-12-31']
```

```python
forex = forex.rename(columns={"Date": "date"})
```

```python
forex = forex.rename(columns={"Value": "FOREX"})
```
